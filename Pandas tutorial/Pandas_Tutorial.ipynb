## Regex Filtering

Get the rows of dataframe where Name **contains** Mega

```python
df.loc[df['Name'].str.contains("Mega")].head()
```

Get the rows of dataframe where Name **does not contain** Mega

```python
df.loc[~df['Name'].str.contains("Mega")].head()
```

Get the rows of dataframe where Type 1 **contains** Fire or Grass

```python
df.loc[df['Type 1'].str.contains('Fire|Grass' , regex = True)]
```

Get the rows of dataframe where Type 1 **contains** Fire or Grass irrespective of case i.e not case-sensitive

```python
df.loc[df['Type 1'].str.contains('fire|grass' , flags = re.I , regex = True)]
```

Get the rows of dataframe where Name **starts with** pi irrespective of case i.e not case-sensitive

```python
df.loc[df['Name'].str.contains('^pi[a-z]*' , flags = re.I , regex = True)]
```

# Conditional Changes

If Type 1 is Fire Make it Flamer

```python
df.loc[df['Type 1'] == 'Fire' , 'Type 1'] = 'Flamer'

df.head()
```

If Type 1 is Flamer Make it Fire

```python
df.loc[df['Type 1'] == 'Flamer' , 'Type 1'] = 'Fire'

df.head()
```

If Type 1 is Fire make Legendary as True

```python
df.loc[df['Type 1'] == 'Fire' , 'Legendary'] = True

df.head()
```

If Total > 500 make Generation and Legendary as TEST VALUE for that row

```python
df.loc[df['Total'] > 500 , ['Generation' , 'Legendary']] = 'TEST VALUE'

df.head()
```

If Total > 500 make Generation as Test 1 and Legendary as Test 2 for that row

```python
df.loc[df['Total'] > 500 , ['Generation' , 'Legendary']] = ['Test 1' , 'Test 2']

df.head()
```

# Aggregate Statistics (Groupby)

```python
df = pd.read_csv('modified.txt' , delimiter = '\t')

df.head()
```

Sort the Type 1 values based on Defense in Ascending order

```python
df.groupby(['Type 1']).mean().sort_values('Defense')
```

Sort the Type 2 values based on Attack in Descending order

```python
df.groupby(['Type 2']).mean().sort_values('Attack' , ascending = False)
```

```python
df.groupby(['Type 1']).sum()
```

Find count of Pokemons based on **count of Type 1**

```python
df['count'] = 1
```

```python
df.groupby(['Type 1']).count()['count']
```

Find count of Pokemons based on **count of Type 1 and Type 2**

```python
df.groupby(['Type 1' , 'Type 2']).count()['count']
```

# Work with Large Amounts of Data

Read in **chunksize** number of rows at a time

```python
new_df = pd.DataFrame(df.columns)

for df in pd.read_csv('modified.csv' , chunksize= 5) : 
  results = df.groupby(['Type 1']).count()
  
  new_df = pd.concat([new_df , results])
  
new_df
```
